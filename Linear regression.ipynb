**Write a code to fit a curve that minimises the squared error cost function using gradient descent
(with learning rate 0.05), on the training set, while the model takes the following form:**

***y = W^T Φn(x), W ∈ R^n+1***

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

# scaler = StandardScaler()
scaler_x = MinMaxScaler(feature_range=(-1,1))
scaler_y = MinMaxScaler(feature_range=(-1,1))

y = df['median_house_value'].values
x = df['median_income'].values
x_reshaped = x.reshape(-1, 1)
y_reshaped = y.reshape(-1, 1)



xs = scaler_x.fit_transform(x_reshaped)
ys = scaler_y.fit_transform(y_reshaped)

x_train, x_test, y_train, y_test = train_test_split(xs.flatten(), ys.flatten(), test_size=0.2, random_state=42)

plt.scatter(x_train, y_train,label="Train")
plt.scatter(x_test, y_test,label="Test")
plt.ylabel("Median House Value")
plt.xlabel("Median_income")
plt.show()
```

```python
def cost(w,x,y):
  predicted=0
  for i in range(len(w)):
    predicted+=w[i]*(x**i)
  cost=predicted-y
  ps=cost**2
  return ps

def meansquarerror(w,x,y,m):
  # x=df[feature]
  # y=df[label]
  totalcost=0
  for i in range(len(x)):
    totalcost+=cost(w,x[i],y[i])
  return totalcost/(2*m)

def predict(w,x):
  predicted=0
  for i in range(len(w)):
    predicted+=w[i]*(x**i)
  return predicted



```

```python
def gradient_decent(w,x_train,y_train,m):
  grad=[0]*(len(w))
  for j in range(900):
      grad=[0]*(len(w))
      # error=0
      for i in range(m):
        p=predict(w,x_train[i])
        e=p-y_train[i]
        for k in range(len(w)):
          grad[k]=grad[k] +  (e* (  (x_train[i])**k)/m  )
        # error+=e*(x_train[i]**index)
      for k in range(len(w)):
        w[k]=w[k]-((0.05)*grad[k])
  return w
```

```python
def plotgraph(w,x_train,y_train) :
    a=list(w)
    a.reverse()
    x_curve = np.linspace(min(x_train), max(x_train))
    y_curve = np.polyval(a, x_curve)
    plt.scatter(x_train, y_train, color='blue', label='Training Data')
    plt.plot(x_curve, y_curve, color='red', label='Polynomial')
    plt.xlabel("Median income(scaled)")
    plt.ylabel("Median house price")
    plt.title("Polynomial vs Training Data")
    plt.legend()
    plt.ylim(-1,1)

    plt.show()
```

```python
def gradient_decent_test(w,x_train,y_train,m):
  grad=[0]*(len(w))
  for j in range(900):
      grad=[0]*(len(w))
      # error=0
      for i in range(m):
        p=predict(w,x_train[i])
        e=p-y_train[i]
        # e=round(e,3)
        for k in range(len(w)):
          # temp=round(x_train[i],3)
          grad[k]=grad[k] +  (e* ( ( x_train[i]**k)/m )  )
        # error+=e*(x_train[i]**index)
      for k in range(len(w)):
        w[k]=w[k]-((0.05)*grad[k])
  return w
```

```python
def testrun(degree):

    m=len(x_train)
    w=[]
    for i in range(degree+1):
      w.append(0)

    w=gradient_decent_test(w,x_train,y_train,m)
    plotgraph(w,x_train,y_train)

    error=meansquarerror(w,x_test,y_test,len(x_test))
    # test_error_list.append(error)
    # error=meansquarerror(w,x_train,y_train,len(x_train))
    # train_error_list.append(error)
    print("Test error",error)
    print(w)
```

### **Regularization**

```python
r = [0, 0.25, 0.5, 0.75, 1]
```

```python
def lasso_reg(w,x_train,y_train,m,r):
  grad=[0]*(len(w))
  for j in range(1000):
      grad=[0]*(len(w))
      # error=0
      for i in range(m):
        p=predict(w,x_train[i])
        e=p-y_train[i]
        for k in range(len(w)):
          grad[k]=grad[k] +  e*(x_train[i]**k)
        # error+=e*(x_train[i]**index)
      for k in range(len(w)):
        x=grad[k]/m
        p=0
        if(w[k]>0):
          p=r
        elif (w[k]<0):
          p=-1*r
        else :
          p=0
        w[k]=w[k]-((0.05)*(x+p))
  return w
```

```python
def ridge_reg(w,x_train,y_train,m,r):
  grad=[0]*(len(w))
  for j in range(1000):
      grad=[0]*(len(w))
      # error=0
      for i in range(m):
        p=predict(w,x_train[i])
        e=p-y_train[i]
        for k in range(len(w)):
          grad[k]=grad[k] +  e*(x_train[i]**k)
        # error+=e*(x_train[i]**index)
      for k in range(len(w)):
        x=grad[k]/m
        p= r*(2*w[k])
        w[k]=w[k]-((0.05)*(x+p))
  return w
```

```python
degree=1
m=len(x_train)
w_lasso=[]
for i in range(degree+1):
  w_lasso.append(0)
# error=meansquarerror(w,x_train,y_train,m)

w_lasso=lasso_reg(w_lasso,x_train,y_train,m,0.5)
ans=[]
for i in w_lasso:
  ans.append(i.item())
print(ans)
n=len(x_test)
error=meansquarerror(w_lasso,x_test,y_test,n)
print("Test error",error)

x = np.linspace(min(x_train), max(x_train))
y =[]
plt.scatter(x_train, y_train, color='blue', label='Training Data')

for i in x:
    y.append(predict(w_lasso,i))

plt.plot(x,y, color='red', label='Fitted Line')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.legend()
plt.grid(True)
plt.ylim(0,550000)
plt.show()
```
